```python
# Function to parse datetime with flexible format handling and debugging
def parse_datetime(date_str):
    try:
        return pd.to_datetime(date_str, errors='coerce', infer_datetime_format=True, dayfirst=True)
    except Exception as e:
        print(f"Error parsing date: {date_str} -> {e}")
        return pd.NaT

# Apply the parsing function to the Date column
df['Date'] = df['Date'].apply(parse_datetime)
```
